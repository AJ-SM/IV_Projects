```python
import cv2 as cv
import numpy as np


# Capturing the image
cap = cv.VideoCapture(0)
# Creating a background Subtractor object and Setting the attributes 
bg_remove = cv.createBackgroundSubtractorMOG2()

# Retaning the Mask For A longer
bg_remove.setHistory(1)

bg_remove.setNMixtures(225)

bg_remove.setDetectShadows(True)

#Detecting The Intensity Of Motion 
bg_remove.setVarThreshold(80)

kernel = None

# Loop for Video Capturing and Displaying
while True:
    ret, frame = cap.read()

    

   

    if not ret:
        break
    main_frame = frame.copy()

    # Bluring the image to restrictring the more detail , to avoid motion noise 
    frame = cv.GaussianBlur(frame,(5,5),0)
        
    # Use To Retaning The Mask For Little Longer Time    
    mask = bg_remove.apply(frame, learningRate=0.005)
    _,mask = cv.threshold(mask,100,255,cv.THRESH_BINARY)

    # Filtring the image to be more clear and able to detect difference between 2 objects clearly
    mask = cv.erode(mask,kernel,iterations = 1 )
    mask = cv.dilate(mask,kernel , iterations =1  )

    
    # Breaking The Mask Into Contour 
    contour, hie = cv.findContours(mask, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_NONE)

    # Loopiong Through Countour Area If its greater certaing value then only the brush will printed 
    for c in contour :
        if cv.contourArea(c) > 5500:

            x,y,width,hight = cv.boundingRect(c)
            cv.rectangle(main_frame,(x,y),(x+width,y+hight),(0,0,225),2)


    cv.imshow('Object Deceted', main_frame)

    # For More Understanding One Can Uncomment this things -- > 

    
    #cv.drawContours(frame, contour, -1, (0, 225, 0))
    #cv.imshow("Mask",mask)
    #cv.imshow("frame",frame)

    if cv.waitKey(1) == ord('q'):
        break

cap.release()
cv.destroyAllWindows()
```
